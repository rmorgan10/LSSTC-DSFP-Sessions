## Problem 1E

Try using larger plotting symbols for the larger stars. Devise a scaling that creates a visual distinction along the main sequence. It doesn’t need to accurately depict differences in stellar radii—just give readers a sense of how radius figures in.

```python
plt.figure()
plt.scatter(df['Teff'].values, 10**df['logLsun'].values, marker='o', 
            facecolor='None', edgecolor='black', s=df['Rsun'].values*2)

plt.xlim(20000, 2000)
plt.yscale('log')
plt.ylabel("Luminosity [L sun]", fontsize=16)
plt.xlabel("Effective Temperature [K]", fontsize=16)
plt.title("Hertzprung-Russell Diagram", fontsize=16)
plt.xticks(fontsize=11)
plt.yticks(fontsize=14)
plt.show()
plt.close()
```

## Problem 1F

Add color. Load the color ramp RdYlBl. Apply it across the plot symbols to match the temperature axis. 

```python
cmap = plt.get_cmap('RdYlBu')
norm = colors.Normalize(vmin=np.log(df['Teff'].values.min()), vmax=np.log(df['Teff'].values.max()))

plt.figure(figsize=(5, 5))
s = plt.scatter(df['Teff'].values, 10**df['logLsun'].values, marker='o', 
            facecolor='None', edgecolor=cmap(norm(np.log(df['Teff'].values))), s=df['Rsun'].values*2,
            cmap='RdYlBl')

plt.xlim(20000, 1000)
plt.ylim(1e-4, 1e5)
plt.yscale('log')
plt.ylabel("Luminosity [L sun]", fontsize=16)
plt.xlabel("Effective Temperature [K]", fontsize=16)
plt.title("Hertzprung-Russell Diagram", fontsize=16)
plt.xticks(fontsize=11)
plt.yticks(fontsize=14)
plt.show()
plt.close()
```
